Give column name as Id, Movie, Year, Rating, Duration

```python
df=pd.read_csv('/content/movies_cleaned.csv',names=['id','Movie','Year','Rating','Duration'])
df
```

Display the number of rows and number of columns

```python
print("Number of rows:", df.shape[0])
print("Number of columns:", df.shape[1])
```

Display the five observations that have highest duration

```python
df.sort_values(by='Duration').tail(5)
```

Remove if any duplicate exists and change the index accordingly

```python
df.reset_index(drop=True,inplace=True)
df
```

Rename the column name â€˜Movieâ€™ as Film

```python
df.rename(columns={'Movie':'Film'},inplace=True)
df
```

Display all the observation starting from index 100 upto 200

```python
df.iloc[100:200]
```

.Display the name and releasing year of the film with maximum duration

```python
df.sort_values(by='Duration',ascending=False)[['Film','Year','Duration']].head(1)
```

.Display the count of films released in each year.

```python
df.groupby('Year').size().reset_index(name='Count')
```

.Find the film released between 1986-1996 having highest duration

```python
df.loc[(df['Year']>=1986) & (df['Year']<=1996)].sort_values(by='Duration',ascending=False)[['Film','Year','Duration']].head(1)
```

Find the films having duration greater than 7000 and rating above 3.5

```python
df.loc[(df['Duration']>7000) & (df['Rating']>3.5)]
```

Print the details about 5 films having less rating , display in the descending order

```python
df.sort_values(by='Rating').head(5)
```

Print the name of the film released in 1996 have highest rating

```python
df.sort_values(by='Rating',ascending=False).loc[df['Year']==1996][['Film','Year','Rating']].head(1)
```
